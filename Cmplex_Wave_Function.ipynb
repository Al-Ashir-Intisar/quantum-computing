```python
# Wave illustrations using complex numbers to create the probability density functions
# Developed by: Dr. Michael P. Haydock - IBM Fellow Emeritus - Data & Optimization Sciences, Inc.
# Initial Coding: 2/11/2025

import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt

# Parameters for the waves
A1 = 1.0  # Amplitude of the first wave
A2 = 1.0  # Amplitude of the second wave
k1 = 1.0  # Wave number of the first wave
k2 = 1.5  # Wave number of the second wave
omega1 = 1.0  # Angular frequency of the first wave
omega2 = 0.5  # Angular frequency of the second wave
phi1 = 0.0  # Phase of the first wave
phi2 = np.pi / 4  # Phase of the second wave

# Define the position and time arrays
x = np.linspace(0, 2 * np.pi, 100)
t = 1.0  # Fixed time for simplicity

# Define the waves using complex exponentials
# Here, 1j represents the imaginary unit . 
# The complex exponential form ** allows us to encode both the amplitude and the phase of the wave in a single expression. 
# This simplifies the mathematics of wave behavior, particularly when dealing with phase differences and superposition.
psi1 = A1 * np.exp(1j * (k1 * x - omega1 * t + phi1))
psi2 = A2 * np.exp(1j * (k2 * x - omega2 * t + phi2))

# Superposition of the waves
# Adding complex exponentials is straightforward because we can simply add the real and imaginary parts. 
# This simplifies the process of finding the resultant wave when two waves interfere. 
# Without complex numbers, we would have to separately handle the real and imaginary parts, making the calculations more cumbersome.
psi = psi1 + psi2
#print(psi)

# Probability density
# The magnitude  of a complex wave function is found using the formula  = sqrt( ), where  is the complex conjugate of . 
# The squared magnitude **2
# represents the probability density, which is a real number. 
# Using complex numbers ensures we correctly account for the phase relationships in the superposed wave.
probability_density = np.abs(psi)**2

# Plot the real part of the wave functions and their superposition
# Real Parts: The plots of the real parts of 1, 2, and their superposition show how the phases and amplitudes combine.
plt.figure(figsize=(12, 6))
plt.plot(x, np.real(psi1), label='Real part of $\psi_1$')
plt.plot(x, np.real(psi2), label='Real part of $\psi_2$')
plt.plot(x, np.real(psi), label='Real part of $\psi_1 + \psi_2$')
plt.xlabel('Position (x)')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Superposition of Two Waves (Real Part)')
plt.grid(True)
plt.show()

# Plot the probability density
# The Probability Density: The plot of the probability density **2
# illustrates the constructive and destructive interference patterns, highlighting how the phases of the waves interact.
plt.figure(figsize=(12, 6))
plt.plot(x, probability_density, label='Probability Density $|\psi|^2$')
plt.xlabel('Position (x)')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Probability Density of Superposed Waves')
plt.grid(True)
plt.show()
```
