```python
# Wave Function Illustration
# Developed by: Dr. Michael P. Haydock - IBM Fellow Emeritus - Data & Optimization Sciences, Inc.
# Initial Coding: 2/11/2025

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Parameters for the wave
A = 1.0  # Amplitude
lambda_ = 2.0  # Wavelength
f = 1.0  # Frequency
phi = 0.0  # Phase

# Derived parameters
k = 2 * np.pi / lambda_
omega = 2 * np.pi * f

# Time and position arrays
x = np.linspace(0, 10, 1000)
t = np.linspace(0, 2, 500)

# Create a figure and axis for the animation
fig, ax = plt.subplots()
line, = ax.plot(x, np.zeros_like(x))

# Function to update the frame
def update(frame):
    y = A * np.sin(k * x - omega * frame + phi)
    line.set_ydata(y)
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=np.linspace(0, 2, 500), blit=True, interval=20)

# Set plot parameters
ax.set_ylim(-A, A)
ax.set_xlabel('Position (x)')
ax.set_ylabel('Wave Displacement (y)')
ax.set_title('Wave Function Animation')

# Display the animation
plt.show()
```
