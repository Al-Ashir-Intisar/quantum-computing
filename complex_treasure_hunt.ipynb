```python
# Complex Number Example – Buried Treasure
# Developed by: Dr. Michael P. Haydock - IBM Fellow Emeritus - Visiting Professor, St. Olaf College
# Initial Coding: 2/19/2025
#
# Description: A neat solution to the following problem uses complex numbers. Think about it:
# An old scroll describes the place where a group of pirates buried a treasure:
# In the island there are only 2 trees, A and B, and the remnants of a gallows. 
# Start at the gallows and count the steps to go in a straight line to tree A. 
# When you arrive, turn 90º to the left and walk the same amount of steps. 
# Where you stopped, place a mark on the ground. 
# Go back to the gallows and count the steps to tree B, also in a straight line.
# When there, turn 90º to the right and walk the same amount of steps forward and place a mark on the ground. 
# Dig in the spot that's halfway through both marks and you'll find the treasure.
# Unfortunately the gallows were no longer there. Find the treasure.

import matplotlib.pyplot as plt

# Coordinates of trees
# Tree A and Tree B as Complex Numbers:
# Let's represent the coordinates of tree A as a complex number 𝑧𝐴=𝑥𝐴+𝑖𝑦𝐴.
# Similarly, represent tree B as 𝑧𝐵=𝑥𝐵+𝑖𝑦𝐵.
# Here, 𝑖 represents the imaginary unit, satisfying 𝑖2=−1.

x_A, y_A = 4, 3
x_B, y_B = -2, 5

# Marking Points with Complex Numbers:
# From the origin (gallows), move to tree A, which is represented by 𝑧𝐴.
# Rotate 90º to the left (multiplying by 𝑖), and walk the same distance to mark the point 𝑀.
# 𝑀=𝑖⋅𝑧𝐴=𝑖(𝑥𝐴+𝑖𝑦𝐴)=𝑖𝑥𝐴−𝑦𝐴=−𝑦𝐴+𝑖𝑥𝐴.

# From the origin (gallows), move to tree B, represented by 𝑧𝐵.
# Rotate 90º to the right (multiplying by −𝑖), and walk the same distance to mark the point 𝑁.
# 𝑁=−𝑖⋅𝑧𝐵=−𝑖(𝑥𝐵+𝑖𝑦𝐵)=−𝑖𝑥𝐵−𝑦𝐵=𝑦𝐵−𝑖𝑥𝐵.
# Coordinates of M and N

M = (-y_A, x_A)
N = (y_B, -x_B)

# Coordinates of the gallows
G = 0,0

# Midpoint of M and N
# Finding the Midpoint: The treasure is buried at the midpoint between points 𝑀 and 𝑁.
# 𝑇=𝑀+𝑁 =(−𝑦𝐴+𝑖𝑥𝐴)+(𝑦𝐵−𝑖𝑥𝐵) =(−𝑦𝐴+𝑦𝐵) + 𝑖(𝑥𝐴−𝑥𝐵)
#   ----  ------------------  ---------  ---------
#    2            2               2          2

T = ((M[0] + N[0]) / 2, (M[1] + N[1]) / 2)
print(T)
print(M[0])
print(M[1])
print(N[0])
print(N[1])
print((M[0] + N[0]) / 2)
print((M[1] + N[1]) / 2)

# Plotting
fig, ax = plt.subplots()
ax.plot([0, x_A], [0, y_A], 'ro-', label='Path to A')
ax.plot([x_A, M[0]], [y_A, M[1]], 'ro:')
ax.plot([0, x_B], [0, y_B], 'bo-', label='Path to B')
ax.plot([x_B, N[0]], [y_B, N[1]], 'bo:')
ax.plot(T[0], T[1], 'go', label='Treasure')

# Adding annotations
# Here's an illustration of the graph with complex numbers: 
# Tree A: 𝑧𝐴=4+3𝑖
# Tree B: 𝑧B=−2+5𝑖

ax.annotate('Tree A (4+3i)', (x_A, y_A))
ax.annotate('Tree B (2+5i)', (x_B, y_B))
ax.annotate('M 90 degrees left -3+4i', M)
ax.annotate('N 90 degrees right 5+2i', N)
ax.annotate('T (Treasure 1+3i)', T)
ax.annotate('   G (gallows)', G)

# Setting labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
ax.grid()
ax.axhline(0, color='black',linewidth=0.5)
ax.axvline(0, color='black',linewidth=0.5)
plt.title('Finding the Buried Treasure  T = ((-3+5)/2) + i((4+2)/2)')

plt.show()
```
